```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 
mpl.rcParams['font.sans-serif'] = ['Times New Roman']
mpl.rcParams['axes.unicode_minus'] = False

# 
kinds_all = ['China MSA', 'MAIB']
kinds = ['S1: Less serious', 'S2: Serious', 'S3: Very serious']
money_scale_0 = [322, 127]  # 1
money_scale_1 = [111, 138, 73]  # 1-
money_scale_2 = [91, 15, 21]  # 2-
dev_position_all = [0.1] * len(kinds_all)
dev_position = [0.1] * len(kinds)  # 

# 
plt.figure(figsize=(8, 8))
# 

# 
ax1 = plt.subplot(211)
wedges1, texts1, autotexts1 = ax1.pie(money_scale_0, autopct='%3.1f%%',  pctdistance=0.7, shadow=True,
                                      explode=dev_position_all, startangle=90,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black','width': 0.5},
                                      colors=plt.cm.Paired(np.linspace(0, 1, len(kinds))))  #  Paired 
ax1.set_title("")
plt.figtext(0.45, 0.53, "Marine accident reports by authority", ha="center", fontsize=14, fontname="Times New Roman")

# 
ax2 = plt.subplot(223)
wedges2, texts2, autotexts2 = ax2.pie(money_scale_1, autopct='%3.1f%%',pctdistance=0.76, shadow=True,
                                      explode=dev_position, startangle=90,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black','width': 0.4},
                                      colors=plt.cm.Set3(np.linspace(0, 1, len(kinds))))  #  Set3 
ax2.set_title("")
plt.figtext(0.20, 0.04, "China MSA: report distribution", ha="center", fontsize=14, fontname="Times New Roman")

# 
ax3 = plt.subplot(224)
wedges3, texts3, autotexts3 = ax3.pie(money_scale_2, autopct='%3.1f%%', pctdistance=0.8,shadow=True,
                                      explode=dev_position, startangle=90,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black', 'width': 0.4},
                                      colors=plt.cm.Set3(np.linspace(0, 1, len(kinds))))  #  Set3 
ax3.set_title("")
plt.figtext(0.7, 0.04, "MAIB: report distribution", ha="center", fontsize=14, fontname="Times New Roman")

# 12
ax1.legend(wedges1, kinds_all, loc='upper right', bbox_to_anchor=(1.25, 1), fontsize=11)
ax2.legend(wedges2, kinds, loc='upper right', bbox_to_anchor=(1.3, 1.2), fontsize=11)
ax3.legend(wedges3, kinds, loc='upper right', bbox_to_anchor=(1.35, 1.2), fontsize=11)

# 
plt.tight_layout()  # 
plt.savefig('-.pdf')
plt.show()
```
