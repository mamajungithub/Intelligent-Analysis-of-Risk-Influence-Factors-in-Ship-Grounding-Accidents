```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 设置字体，避免中文乱码
mpl.rcParams['font.sans-serif'] = ['Times New Roman']
mpl.rcParams['axes.unicode_minus'] = False

# 数据
kinds_all = ['China MSA', 'MAIB']
kinds = ['S1: Less serious', 'S2: Serious', 'S3: Very serious']
money_scale_0 = [322, 127]  # 图1数据
money_scale_1 = [111, 138, 73]  # 图1数据-中国海事局
money_scale_2 = [91, 15, 21]  # 图2数据-英国海事局
dev_position_all = [0.1] * len(kinds_all)
dev_position = [0.1] * len(kinds)  # 扇区分离程度

# 创建图表
plt.figure(figsize=(8, 8))
# 公共参数设置

# 绘制第一个子图
ax1 = plt.subplot(211)
wedges1, texts1, autotexts1 = ax1.pie(money_scale_0, autopct='%3.1f%%',  pctdistance=0.7, shadow=True,
                                      explode=dev_position_all, startangle=90,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black','width': 0.5},
                                      colors=plt.cm.Paired(np.linspace(0, 1, len(kinds))))  # 使用 Paired 渐变色
ax1.set_title("")
plt.figtext(0.45, 0.53, "Marine accident reports by authority", ha="center", fontsize=14, fontname="Times New Roman")

# 绘制第二个子图
ax2 = plt.subplot(223)
wedges2, texts2, autotexts2 = ax2.pie(money_scale_1, autopct='%3.1f%%',pctdistance=0.76, shadow=True,
                                      explode=dev_position, startangle=90,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black','width': 0.4},
                                      colors=plt.cm.Set3(np.linspace(0, 1, len(kinds))))  # 使用 Set3 渐变色
ax2.set_title("")
plt.figtext(0.20, 0.04, "China MSA: report distribution", ha="center", fontsize=14, fontname="Times New Roman")

# 绘制第三个子图
ax3 = plt.subplot(224)
wedges3, texts3, autotexts3 = ax3.pie(money_scale_2, autopct='%3.1f%%', pctdistance=0.8,shadow=True,
                                      explode=dev_position, startangle=90,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black', 'width': 0.4},
                                      colors=plt.cm.Set3(np.linspace(0, 1, len(kinds))))  # 使用 Set3 渐变色
ax3.set_title("")
plt.figtext(0.7, 0.04, "MAIB: report distribution", ha="center", fontsize=14, fontname="Times New Roman")

# 添加图例，设置字体大小为12号
ax1.legend(wedges1, kinds_all, loc='upper right', bbox_to_anchor=(1.25, 1), fontsize=11)
ax2.legend(wedges2, kinds, loc='upper right', bbox_to_anchor=(1.3, 1.2), fontsize=11)
ax3.legend(wedges3, kinds, loc='upper right', bbox_to_anchor=(1.35, 1.2), fontsize=11)

# 设置整体布局
plt.tight_layout()  # 自动调整子图参数，避免压盖
plt.savefig('贝叶斯-数据集分类.pdf')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 设置字体，避免中文乱码
mpl.rcParams['font.sans-serif'] = ['Times New Roman']
mpl.rcParams['axes.unicode_minus'] = False


# ========== 颜色方案优化 ==========
# 主色系（使用HSL协调色相）
main_colors = [
    '#4E79A7',  # 深蓝
    '#F28E2B',  # 橙
    '#E15759',  # 红
    '#76B7B2',  # 青
    '#59A14F',  # 绿
    '#EDC948',  # 黄
    '#B07AA1',  # 紫
    '#FF9DA7',  # 粉
    '#9C755F',  # 棕
    '#BAB0AC'   # 灰
]

# 辅助色（主色的低饱和度版本）
pastel_colors = [
    '#AEC7E8',  # 浅蓝
    '#FFD8A8',  # 浅橙
    '#FFB3B3',  # 浅红
    '#C7E5DB',  # 浅青
    '#C4E0B4',  # 浅绿
    '#FFEBA8',  # 浅黄
    '#D4BBDD',  # 浅紫
    '#FFD1D6',  # 浅粉
    '#D7C4B8',  # 浅棕
    '#E0DEDB'   # 浅灰
]
# 数据
kinds_all = ['China MSA', 'MAIB']
kinds = ['S$_1$: Less serious', 'S$_2$: Serious', 'S$_3$: Very serious']
money_scale_0 = [46, 148]  # 图1数据
money_scale_1 = [1, 31, 14]  # 图1数据-中国海事局
money_scale_2 = [72, 49, 26]  # 图2数据-英国海事局
dev_position_all = [0.1] * len(kinds_all)
dev_position = [0.1] * len(kinds)  # 扇区分离程度

# 创建图表
plt.figure(figsize=(8, 8))
# 公共参数设置

# 绘制第一个子图
ax1 = plt.subplot(211)
wedges1, texts1, autotexts1 = ax1.pie(money_scale_0, autopct='%3.1f%%',  pctdistance=0.7, shadow=True,
                                      explode=dev_position_all, startangle=0,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black','width': 0.5},
                                      colors=main_colors[:len(kinds_all)])  # 动态取色  # 使用 Paired 渐变色
ax1.set_title("")
plt.figtext(0.45, 0.53, "Ship grounding accident reports by authority", ha="center", fontsize=14, fontname="Times New Roman")

# 绘制第二个子图
ax2 = plt.subplot(223)
wedges2, texts2, autotexts2 = ax2.pie(money_scale_1, autopct='%3.1f%%',pctdistance=0.76, shadow=True,
                                      explode=dev_position, startangle=90,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black','width': 0.4},
                                      colors=pastel_colors[:len(kinds)]) # 与主色对应
ax2.set_title("")
plt.figtext(0.20, 0.04, "China MSA: reports distribution by S$_i$", ha="center", fontsize=14, fontname="Times New Roman")

# 绘制第三个子图
ax3 = plt.subplot(224)
wedges3, texts3, autotexts3 = ax3.pie(money_scale_2, autopct='%3.1f%%', pctdistance=0.8,shadow=True,
                                      explode=dev_position, startangle=90,
                                      wedgeprops={'linewidth': 1, 'edgecolor': 'black', 'width': 0.4},
                                       colors=pastel_colors[:len(kinds)]) # 与主色对应  # 延续辅助色  # 使用 Set3 渐变色
ax3.set_title("")
plt.figtext(0.7, 0.04, "MAIB: reports distribution by S$_i$", ha="center", fontsize=14, fontname="Times New Roman")

# 添加图例，设置字体大小为12号
ax1.legend(wedges1, kinds_all, loc='upper right', bbox_to_anchor=(1.25, 1), fontsize=11)
ax2.legend(wedges2, kinds, loc='upper right', bbox_to_anchor=(1.3, 1.2), fontsize=11)
ax3.legend(wedges3, kinds, loc='upper right', bbox_to_anchor=(1.35, 1.2), fontsize=11)

# 设置整体布局
plt.tight_layout()  # 自动调整子图参数，避免压盖
plt.savefig('搁浅-数据集分类.pdf')
plt.show()
```
